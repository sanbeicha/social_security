```python
data_15 = pd.read_excel('/Users/tanping/Desktop/09-15-1.xls',header=None,names=['地区','城镇就业人员年末数','年末职工参保人数（万人）','基金收入','基金支出','累计结余'])
data_15.head()
```

```python
from pyecharts import Bar 
attr=data_15['地区']
v1=data_15['城镇就业人员年末数']
v4=data_15['年末职工参保人数（万人）']  
bar=Bar("90-16年人数变化情况")  
bar.add('城镇就业人员年末数',attr,v1)  #标记点
bar.add('年末职工参保人数（万人）',attr,v4) #标记线
bar
```

```python
from pyecharts import Line,EffectScatter,Overlap
attr = data_15['地区']
v1=data_15['城镇就业人员年末数']
line = Line('城镇就业人员年末数闪烁图示例')
line.add('',attr,v1,is_random = True)
es = EffectScatter()
es.add('',attr,v1,effect_scale=8)   #闪烁
overlop = Overlap()
overlop.add(line)                   
overlop.add(es)
overlop
```

```python
attr = data_15['地区']
v4=data_15['年末职工参保人数（万人）'] 
line = Line('年末职工参保人数闪烁图示例')
line.add('',attr,v4,is_random = True)
es = EffectScatter()
es.add('',attr,v4,effect_scale=8)   #闪烁
overlop = Overlap()
overlop.add(line)                   
overlop.add(es)
overlop
```
